## 1. Grounding in Reality: The On-chain Anchor

Our first objective was to define a "True Value" anchor to avoid buying into market euphoria. We focused on the **MVRV Ratio** (Market Value to Realized Value), which essentially compares the current market price to the average cost basis of all holders.

## 2. Searching for Divergence: The Z-Score Lens

Raw prices often hide true momentum. By standardizing diverse metrics like **HashRate**, **Price**, and **MVRV** into Z-Scores, we can spot fundamental front-running.

<img src='figures/standardized_comparison.png' width='800'>

**Strategic Observation:** "Network Adoption vs. Price Lag". Notice periods where **HashRate Z-Scores** (representing network security and miner investment) trend upward while **Price Z-Scores** remain flat. These divergences often precede major bullish reversals, providing a 'fundamental' green light for our strategy.

## 3. The Sentiment Breakthrough: Can Prediction Markets See the Future?

The pinnacle of this exploration was testing the predictive power of Polymarket. We asked: *"Do people bet on Bitcoin's future before the price actually moves?"*

The answer is a statistically significant **YES**.

<img src='figures/lead_lag_correlation.png' width='800'>

**Discovery: The 5-10 Day Sentiment Lead**
- **Negative Lag Peak:** Cross-correlation analysis shows that activity in Polymarket (market creation and volume) peaks **5 to 10 days BEFORE** the actual price realization in spot markets.
- **Why it matters:** Prediction markets act as an aggregation of 'Skin-in-the-game' intelligence. Spikes in Poly-activity serve as an early-warning radar for our DCA model to adjust buying weights before the volatility hits the main trading charts.

## 4. Conclusion & Final Strategic Blueprint

This EDA has transformed raw data into a three-layered trading sieve:
1.  **On-chain Value (MVRV):** Tells us *what* to buy (under-valuation detection).
2.  **Risk Management (Drawdown):** Tells us *how much* pain we are in.
3.  **Sentiment Lead (Polymarket):** Provides a **5-10 day window** to front-run spot market momentum.

**Final Verdict:** We will proceed with a **Hybrid DCA Model** that weights daily purchases based on these discovered correlations, specifically prioritizing the Polymarket Lead-Lag signal for short-term tactical adjustments.
