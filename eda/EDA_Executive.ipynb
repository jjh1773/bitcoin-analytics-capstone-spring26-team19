# 📊 Strategic Bitcoin EDA: On-chain & Sentiment Fusion

**Objective:** To systematically evaluate Bitcoin's value cycles by merging **On-chain Fundamentals** with **Prediction Market Intelligence**. This report outlines key discoveries that drive our algorithmic DCA strategy.

---

# 1. Executive Summary

This analysis identifies the precise relationship between Bitcoin's network value and global crowd sentiment captured via Polymarket. 

**Key Discoveries:**
- **The 5-10 Day Sentiment Lead:** Polymarket activity (volume/count) consistently leads Bitcoin price action by 5 to 10 days, providing a tactical window for front-running spot market momentum.
- **MVRV Value Floor:** The MVRV Ratio effectively defines the 'pain floor' (at 1.0) and 'euphoria ceiling' (above 3.0), serving as a macro compass for our DCA weightings.
- **Hybrid Strategy:** By combining these two independent high-signal data sources, we can transition from a passive DCA model to a dynamic intelligence-led accumulation strategy.

# 2. Data Retrieval & Preprocessing

Our pipeline integrates diverse datasets to capture a holistic view of the market:

### **A. Data Sources**
- **Coin Metrics (On-chain/Price):** Daily BTC spot price, Market Cap, Realized Cap, and MVRV Z-Scores.
- **Polymarket (Sentiment/Volume):** High-frequency event betting data (Finance/Politics categories) processed into daily aggregates.

### **B. Methods & Preprocessing**
- **Smoothing:** Applied **7-day rolling means** to mitigate weekend volatility in betting activity.
- **Normalization:** Transformed raw metrics into **Z-Scores** to enable cross-variable comparison between price action and hash rate usage.
- **Alignment:** Merged datasets via daily UTC timestamps, ensuring look-ahead bias is strictly avoided.
