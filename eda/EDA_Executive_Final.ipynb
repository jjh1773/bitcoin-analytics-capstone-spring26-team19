## **Layer 1: Strategic Insight & Strategy Evaluation**

---

### **1. Problem & Angle**

**Objective:** To improve Bitcoin accumulation efficiency by transitioning from passive DCA to an intelligence-led dynamic model. 

**The Angle:** We hypothesize that **Prediction Market Data (Polymarket)** acts as a "Skin-in-the-game" leading indicator for global sentiment, which precedes spot market price action. By fusing this tactical sentiment with **On-chain Fundamentals (MVRV)**, we aim to outpace a uniform DCA benchmark.

### **2. Core Insight**

**The 5-10 Day Sentiment Lead:** Through rigorous cross-correlation analysis between Polymarket betting volume/market count and BTC spot prices, we discovered that prediction market activity consistently **leads Bitcoin price movements by 5 to 10 days**.

This provides a strategic window for tactical weight adjustments—allowing us to increase accumulation before a sentiment-driven breakout and reduce during euphoria.

### **4. Interpretation & Strategy Design**

**The Hybrid Accumulation Strategy:**
We implemented a weighted multiplier model:
- **Primary (64%):** MVRV Value (Macro cycle floor)
- **Secondary (16%):** 200-day Moving Average (Trend confirmation)
- **Tertiary (20%):** Polymarket Sentiment (Short-term tactical lead)

This allows for high conviction during cyclical bottoms while keeping a 'tactical radar' active for short-term sentiment shifts.

### **5. Limitations**

- **Data Granularity:** Polymarket data is high-fidelity but carries noise during lower-volume periods.
- **Market Maturation:** As Prediction Markets mature, the lead time may shrink as the signal becomes more widely exploited (alpha decay).
- **Assumptions:** We assume 'BTC-related' market questions are the primary proxies for crypto sentiment.
